```python
import pandas as pd

df = pd.read_csv('iris_clean.csv')
df
```

```python
x = df.drop('species', axis=1)
x
```

```python
y = df['species']
y
```

```python
x = df.drop('species', axis=1)
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression

# Instantiate the model
lr = LogisticRegression()

# Train the model
lr.fit(x_train, y_train)
```

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
print(y_lr_train_pred,y_lr_test_pred)
```

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Assuming y_lr_test_pred and y_lr_train_pred are the predicted labels
# Ensure y_train and y_test are the true labels

# Calculate accuracy
accuracy_train = accuracy_score(y_train, y_lr_train_pred)
accuracy_test = accuracy_score(y_test, y_lr_test_pred)

# Calculate precision
precision_score_train = precision_score(y_train, y_lr_train_pred, average='weighted')
precision_score_test = precision_score(y_test, y_lr_test_pred, average='weighted')

# Calculate recall
recall_score_train = recall_score(y_train, y_lr_train_pred, average='weighted')
recall_score_test = recall_score(y_test, y_lr_test_pred, average='weighted')

# Calculate F1-score
f1_score_train = f1_score(y_train, y_lr_train_pred, average='weighted')
f1_score_test = f1_score(y_test, y_lr_test_pred, average='weighted')
```

```python
lr_results = pd.DataFrame(['Linear regression', accuracy_train,accuracy_test,precision_score_train,precision_score_test,recall_score_train,recall_score_test,f1_score_train,f1_score_test]).transpose()
lr_results.columns = ["Method","Accuracy Train","Accuracy Test","Precision Train","Precision Test","Recall Train","Recall Test","F1 Train","F1 Test"]
```

```python
lr_results
```
